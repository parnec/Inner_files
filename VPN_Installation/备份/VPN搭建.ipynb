# 搭建VPN

## ssh连接VPS

```bash
ssh root@你的VPS地址
```

## 开始搭建

**Note:**
推荐第一种，即 Ubuntu环境 。

### Ubuntu环境

在远程服务器端安装shadowsocks

```bash
apt-get install python-pip
pip install shadowsocks
```

再用nano新建配置文件

```bash
nano /etc/shadowsocks.json
```

输入以下内容


    { 
      "server":"你的VPS地址", 
      "server_port":你的VPS端口号, 
      "local_address": "127.0.0.1", 
      "local_port":1080, 
      "password":"你的VPN密码",
      "timeout":300, 
      "method":"aes-256-cfb", 
      "fast_open": false } 




### Centos环境

下载脚本

```bash
wget -O ss.sh http://zhangdanyang.com/file/ss.sh
```

**Note:**
备份文件夹里保留有 ss.sh ，当网上资源下载不了时可以直接用里面的。

执行脚本

```bash
bash ss.sh
```

# 配置加速器


## 选择一种加速器

**Note:**
推荐第一种，即 bbr加速器 。

### bbr加速器

```bash
wget –no-check-certificate https://github.com/teddysun/across/raw/master/bbr.sh
chmod +x bbr.sh
./bbr.sh
```

**Note:**
备份文件夹里保留有 bbr.sh ，当网上资源下载不了时，可以直接用里面的。

### 锐速加速器(只适用于Centos6)

```bash
wget http://ftp.scientificlinux.org/linux/scientific/7.4/x86_64/updates/security/kernel-3.10.0-693.2.2.el7.x86_64.rpm
rpm -ivh kernel-3.10.0-693.2.2.el7.x86_64.rpm --force
wget -N --no-check-certificate https://raw.githubusercontent.com/91yun/serverspeeder/master/serverspeeder-all.sh && bash serverspeeder-all.sh
```

## 剩余操作

重启

```bash
reboot
```

ssh连接到VPS

```bash
ssh root@你的VPS地址
```

激活shadowsocks
```bash
nohup ssserver -c /etc/shadowsocks.json &
```
